```python
df_shopping["Review Rating"] = df_shopping.groupby("Category")["Review Rating"].transform(lambda x: x.fillna(x.median()))
```

```python
df_shopping.isnull().sum()
```

#Transformando as colunas em minusculas e tirando os espaos em branco

```python
df_shopping.columns = df_shopping.columns.str.lower()
df_shopping.columns = df_shopping.columns.str.replace(" ", "_")
df_shopping.rename(columns={"purchase_amount_(usd)":"purchase_amount"}, inplace=True)
```

```python
df_shopping.columns
```

#Criando uma coluna de age_group para agrupar os clientes em 4 faixa-etaria

```python
labels = ["Jovem", "Adulto", "Meia-Idade", "Idoso"]
df_shopping["age_group"] = pd.qcut(df_shopping["age"], q=4, labels=labels)
```

```python
df_shopping[["age", "age_group"]].head(10)
```

#Criando uma coluna de purchase_frequency_days para saber a frequencia que o cliente compra (semanal, mensal, bimestral ...)

```python
frequency_mapping = {
    'Weekly': 7,
    'Biweekly': 14,
    'Fortnightly': 14,
    'Monthly': 30,
    'Quaterly': 90,
    'Every 3 months': 90,
    "Annually": 365
}

df_shopping['purchase_frequency_days'] = df_shopping['frequency_of_purchases'].map(frequency_mapping).fillna(0).astype(int)
```

```python
df_shopping[['purchase_frequency_days', 'frequency_of_purchases']].head(20)
```
